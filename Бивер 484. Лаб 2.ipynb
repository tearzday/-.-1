1. Поменяйте цвета своих изображений на противоположныев RGB используя прямые вычисления значений для каждого из каналов.


```python
import random
from PIL import Image , ImageDraw , ImageOps # Подключим необходимые библиотеки
from matplotlib import pyplot as plt
img0 = Image.open("res.jpg") #Открываем изображение
img1 = Image.open("ris1.jpg") #Открываем изображение
img2 = Image.open("ris2.jpg") #Открываем изображение
img3 = Image.open("ris3.jpg") #Открываем изображение

def draw(img):
  draw = ImageDraw.Draw(img) #Создаем инструмент для рисования
  width = img.size[0] #Определяем ширину
  height = img.size[1] #Определяем высоту
  pix = img.load() #Выгружаем значения пикселей
  draw = ImageDraw.Draw(img ) #Создаем инструмент для рисования
  for i in range(width):
    for j in range(height):
      r = pix[i, j][0] #попиксельно проходимся по всей картинке
      g = pix[i, j][1] 
      b = pix[i, j][2] 
      draw.point((i, j), (255 - r, 255 - g, 255- b)) #(255 - r, 255 - g, 255- b) используется для вычисления из получнных пикселей 255 , тем самым получаем негатив
  img.save('result1.jpg')
##Вывод получившихся изображений
  plt.imshow( Image.open('result1.jpg'))
  plt.show()


draw(img0)
draw(img1)
draw(img2)
draw(img3)
```

2. Получите инвертированные по цвету изображения с использованиемметода invert из модуля ImageOps библиотеки PIL.

```python

from PIL import Image , ImageDraw,ImageOps # Подключим необходимые библиотеки
from matplotlib import pyplot as plt

img0 = Image.open("res.jpg") #Открываем изображение
img1 = Image.open("ris1.jpg") #Открываем изображение
img2 = Image.open("ris2.jpg") #Открываем изображение
img3 = Image.open("ris3.jpg") #Открываем изображение
def invert(im):
  ##В модуле ImageOps есть встроенная реализация превращения в негатив - фунцикя invert
  im = ImageOps.invert(im)
  im.save('result2.jpg')
  ##Выводим результат применения функции
  plt.imshow( Image.open('result2.jpg'))
  plt.show()

 


invert(img0)
invert(img1)
invert(img2)
invert(img3)
```

**Градация серого**

```python

from PIL import ImageOps # Подключим необходимые библиотеки
img0 = Image.open("res.jpg") #Открываем изображение
img1 = Image.open("ris1.jpg") #Открываем изображение
img2 = Image.open("ris2.jpg") #Открываем изображение
img3 = Image.open("ris3.jpg") #Открываем изображение
def draw(im):

  ##Градация серого
  im3 = ImageOps.grayscale(im) #grayscale является встроенной функции для получения градации серого
  im3.save('gray.jpg')
  ##Выводим результат получившейся градации серого
  plt.imshow(Image.open('gray.jpg'))
  plt.show()


draw(img0)
draw(img1)
draw(img2)
draw(img3)
```

**colorize**

```python
def draw(im):
  im3 = ImageOps.grayscale(im)
  ##colorize является встроенной функции для получения градации указанных нами цветов
  ##В нашем случае , это yellow - желтый и purlpe - пурпурный
  im4 = ImageOps.colorize(im3,'yellow', 'purple', mid = None,blackpoint = 0, whitepoint = 255, midpoint = 127)
  im4.save('color.jpg')
  ##Выводим результат получившейся градации
  plt.imshow(Image.open('color.jpg'))
  plt.show()

draw(img0)
draw(img1)
draw(img2)
draw(img3)
```

4. Используя метод transpose, поверните свои изображения на90градусовпротив часовой стрелки.

```python
from PIL import ImageOps # Подключим необходимые библиотеки
img0 = Image.open("res.jpg") #Открываем изображение
img1 = Image.open("ris1.jpg") #Открываем изображение
img2 = Image.open("ris2.jpg") #Открываем изображение
img3 = Image.open("ris3.jpg") #Открываем изображение
def draw(im):

  im = im.transpose(Image.ROTATE_270)##Встроенная функция transpose со значение в 90 градусов. Так как Rotate поворачивает против часовой стрелки , то указываем 270 градусов
  im.save('result4.jpg')
  ##Вывод результата
  plt.imshow( Image.open('result4.jpg'))
  plt.show()


draw(img0)
draw(img1)
draw(img2)
draw(img3)
```

5. Создайте bmp-изображения с 8, 12, 16 и 24 цветами из ранееполученныхкартинок.

```python

import random
from PIL import Image , ImageDraw , ImageOps # Подключим необходимые библиотеки
from matplotlib import pyplot as plt
img0 = Image.open("res.jpg") #Открываем изображение
img1 = Image.open("ris1.jpg") #Открываем изображение
img2 = Image.open("ris2.jpg") #Открываем изображение
img3 = Image.open("ris3.jpg") #Открываем изображение
def bmp8(im):
  ##quantize для сокращения цветов в палитре
  im8 = im.quantize(8)
  im8.save('b8.bmp','BMP')
  return (Image.open('b8.bmp'))

def bmp12(im):
  im12 = im.quantize(12)
  im12.save('b12.bmp','BMP')
  return (Image.open('b12.bmp'))

def bmp16(im):
  im16 = im.quantize(16)
  im16.save('b16.bmp','BMP')
  return (Image.open('b16.bmp')) 

def bmp24(im):  
  im24 = im.quantize(24)
  im24.save('b24.bmp','BMP')
  return (Image.open('b24.bmp'))

def draw2(image1):
 plt.figure(figsize=(12,7), dpi=90)
 plt.subplot(1, 4, 1)
 plt.imshow(bmp8(image1))
 plt.subplot(1, 4, 2)
 plt.imshow(bmp12(image1))
 plt.subplot(1, 4, 3)
 plt.imshow(bmp16(image1))
 plt.subplot(1, 4, 4)
 plt.imshow(bmp24(image1))
 plt.show()


draw2(img0)
draw2(img1)
draw2(img2)
draw2(img3)
```
